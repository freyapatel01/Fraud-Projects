Extract Month, Week, Quarter and Year data from Transaction Date to leverage them for aggregation

```python
credit_card['Posted Date'] = pd.to_datetime(credit_card['Posted Date'], errors = 'coerce')
credit_card['Transaction Date'] = pd.to_datetime(credit_card['Transaction Date'], errors = 'coerce')
```

```python
credit_card['Trans_Week'] = credit_card['Transaction Date'].dt.isocalendar().week
credit_card['Trans_Year']  = credit_card['Transaction Date'].dt.isocalendar().year
credit_card['Trans_Month']  = credit_card['Transaction Date'].dt.month
credit_card['Transaction_Quarter'] = credit_card['Transaction Date'].dt.quarter
credit_card['Day Of Week'] = credit_card['Transaction Date'].dt.isocalendar().day
```

1. By comparing each transaction amount to the category (Agency & Merchant's) median value we have rough understanding on how much each transaction compares to the median value. A high ratio (let's say double) would be flagged out as an anomaly. Or consistent high ratios would indicate overuse.

```python
credit_card['Median_Amount_Ratio'] = credit_card['Amount']-credit_card['Trans_median']
```

2. Number of transactions in a day by cardholder. Keeping track of cardholder's spending habits by counting the number of transactions in a day helps flag unusually high number. 
For instance if the number is above the 75th percentile on a day then this would be an outlier worth noting.

```python

credit_card = credit_card.sort_values(by=['Transaction Date', 'Cardholder Last Name'])

Trans_in_a_day = credit_card.groupby(['Cardholder Last Name','Transaction Date']).size()
Trans_in_a_day = pd.DataFrame(Trans_in_a_day)
Trans_in_a_day.columns = ['Trans_in_a_day']
```

```python
credit_card = pd.merge(credit_card, Trans_in_a_day,how = 'left', on = ['Transaction Date', 'Cardholder Last Name'])
```

5. Number of transactions by each user in each week of the year.
Feature: Ratio of number of transactions by a cardholder in a week and total number by the agency in that week. This ratio helps understand what percent of agency's total transactions are accounted for by each individual. Such ratio's help flag unually high proportions and thus fraudulent card use.

```python
Agency_Counts = credit_card[['Agency Name','Trans_Week','Trans_Year']].value_counts()
```

```python
Agency_Individual_Counts = credit_card[['Agency Name','Cardholder First Initial','Cardholder Last Name','Trans_Week','Trans_Year']].value_counts()
```

```python
Prop_Df = pd.DataFrame(Proportion_of_use_weekly)
```

```python
Prop_Df.columns = ['Prop of Transactions']
```

```python
credit_card = pd.merge(credit_card,Prop_Df,how = 'left', on = ['Agency Name','Cardholder First Initial','Cardholder Last Name','Trans_Week','Trans_Year'])
```

6. Z_Score Calculation to view the difference between Transaction Amounts and Mean in terms of Standard Deviation. The Z-Score is an excellent identifier of anomalies as it helps understand the distance from mean. A general rule of thumb is that a value outside +3 and -3 would be considered an outlier.

```python
credit_card['Individual_Z_Score'] = (credit_card.Amount - credit_card.Trans_mean)/credit_card.Trans_Std
```

7. Days from Last Transaction. This helps in understanding the recency aspect of transactions. If the Days since last transaction stays 0 consistently for an individual versus the agency in general then, it could be a sign of overuse. 

```python
df_days = credit_card.sort_values(['Cardholder Last Name','Transaction Date'])
df_days = credit_card.groupby(['Cardholder Last Name'])['Transaction Date'].diff().fillna(0)
```

```python
df_days.name = 'Days_From_Last_Cardholder_Transaction'
```

```python
credit_card = credit_card.merge(df_days, how='left', left_index=True, right_index=True)
```

```python
credit_card['Days_From_Last_Cardholder_Transaction'] = credit_card['Days_From_Last_Cardholder_Transaction'].apply(lambda x: x.days if x!=0 else 0)
```

9. Large Purchase Threshold which flags purchases above the 90th percentile. When a transaction crosses 90th percentile of all transaction data, that would be considered an 
outlier. The Large Purchase threshold provides a benchmark value for each agency and MCC. Next, with a boolean value we check whether an amount is above or below this high value.
Consistent 1's for a cardholder would help understand overspending trends.

```python

large_purchase_threshold = credit_card.groupby(['Agency Name','Merchant Category Code (MCC)'])['Amount'].quantile(0.90)
```

```python
large_purchase_threshold = pd.DataFrame(large_purchase_threshold)
```

```python
large_purchase_threshold.columns = ['Large_Purchase']
```

```python
large_purchase_threshold
```

```python
credit_card = pd.merge(credit_card,large_purchase_threshold, how = 'left', 
         on = ['Agency Name','Merchant Category Code (MCC)'])
```

```python
# Create a binary feature indicating whether a transaction is a large purchase

credit_card['Is Large Purchase'] = np.where(credit_card.Amount >= credit_card['Large_Purchase'],1,0)
```

11. Moving Average Over a 10 day period. This helps understand how transaction trends track over a period of 10 days: aggregated over 'Agency Name','Merchant Category Code (MCC)','Transaction Date'. A 10 day window allows for a smoother curve and looser fit, thus helping us identify stark outliers.  

```python
credit_card = credit_card.sort_values(by=['Transaction Date','Agency Name','Merchant Category Code (MCC)'])
Trans_Date_Group = credit_card.groupby(['Agency Name','Merchant Category Code (MCC)','Transaction Date'])['Amount'].mean()
Trans_Date_Group = pd.DataFrame(Trans_Date_Group)
Trans_Date_Group.columns = ['Trans_Date_Group']
```

```python
credit_card = pd.merge(credit_card, Trans_Date_Group,how = 'left', on = ['Agency Name','Merchant Category Code (MCC)','Transaction Date'])
```

```python
credit_card = credit_card.sort_values(by=['Agency Name', 'Merchant Category Code (MCC)', 'Transaction Date'])
```

```python
credit_card['rolling_mean_10'] = credit_card.groupby(['Agency Name', 'Merchant Category Code (MCC)'])['Amount'].transform(lambda x: x.rolling(window=10).mean())
```

12 & 13. Rolling Min and Max Transaction Amount by Agency and Vendor aggregated over a 4 week period. Spends change with time and so do their max and min. Thus, a 4 week window is not too small (to avoid volatility), nor too big to avoid missing out on building trends. A sudden spike or drop outside the rolling min and max would be cause for action.

```python
credit_card['Rolling_Max_4'] = credit_card.groupby(['Agency Name','Merchant Category Code (MCC)','Trans_Week','Trans_Year'])['Amount'].transform(lambda x: x.rolling(window=4).max())
```

```python
credit_card['Rolling_Min_4'] = credit_card.groupby(['Agency Name','Merchant Category Code (MCC)','Trans_Week','Trans_Year'])['Amount'].transform(lambda x: x.rolling(window=4).min())
```
